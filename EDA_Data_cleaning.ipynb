```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df.columns
```

```python
df.dtypes
```

```python
#Null values for each column 
df.isnull().sum()
```

```python
#Handling null values of age column
mean = round(df['Age'].mean(), 2)
```

```python
df['Age'] = df['Age'].fillna(mean)
```

```python
#Dropping unneccesary columns 
df.drop('Cabin', axis = 1, inplace = True)
df.drop('PassengerId', axis = 1, inplace = True)
```

```python
#Filling null value in 'Embarked' column with most repeated values 
mode = df['Embarked'].mode()
df['Embarked'] = df['Embarked'].fillna(str(mode))
```

```python
#Null values are handled
df.isnull().sum()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'Pclass')
```

```python
#Outliers
sns.boxplot(data=df, palette = 'rainbow', orient = 'h')
```

```python
#Correlation between two variables
correlation = df[['Age', 'Survived']]
```

```python
correlation.corr()
```

```python
#No of female and male passengers
sns.countplot(x = 'Sex', data=df)
```
